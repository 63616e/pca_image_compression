```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import numpy as np
```

```python
# convert image to grayscale 
grayslan = Image.open('image.jpg').convert('L')
plt.imshow(grayslan)
```

```python
from sklearn.decomposition import PCA, IncrementalPCA

pca = PCA()

pca.fit(grayslan)
# need to find cumsum to know where we keep 95% of the variance or "originality"

best_k = np.argmax(np.cumsum(pca.explained_variance_ratio_) >=0.95)
print(best_k)
# aslan is kept at his best shape when we have 11 k components
```

```python
ipca = IncrementalPCA(n_components=best_k)
image_recon = ipca.inverse_transform(ipca.fit_transform(grayslan))

# Plotting the reconstructed image
plt.figure(figsize=[12,8])
plt.imshow(image_recon,cmap = plt.cm.gray)

# how the fuck is this a good image
```

```python
ks = [5,10,15,20,25,30]

plt.figure(figsize=[15,9])

for i in range(6):
    plt.subplot(2,3,i+1)
    ipca = IncrementalPCA(n_components=ks[i])
    image_recon = ipca.inverse_transform(ipca.fit_transform(grayslan))
    new_p = Image.fromarray(image_recon).convert('RGB').save('exported_images/image{0}.jpeg'.format(ks[i])) # save the image
    plt.imshow(image_recon,cmap = plt.cm.gray) # plot the image
    plt.title("Components: "+str(ks[i])) # add title to each subplot
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```
